```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,10, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Object Oriented Approach
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])#Range between 0 - 1
axes.plot(x,y)
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Title')

```

```python
#Mutiple Axes to understand more
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,1,1])
axes2 = fig.add_axes([0.3,0.5,0.3,0.3])
```

```python
#Plotting with multiple axes
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,1,1])
axes2 = fig.add_axes([0.3,0.5,0.3,0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x,'b')

axes1.set_title('Outer Plot')
axes2.set_title('Inner Plot')
```
