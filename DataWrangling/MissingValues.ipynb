```python
# Load library
import pandas as pd
```

```python
# File URL
url = 'titanic.csv'
```

```python
# Load data
dataframe = pd.read_csv(url)
```

```python
#Display data
dataframe.head(10)
```

```python
#Drop rows with NAN
dataframe.dropna()
```

```python
#Drop columns with NAN(Age has NaN - So it is removed)
dataframe.dropna(axis=1)
```

```python
#Fill the NaN age with its mean value
dataframe.fillna(value=dataframe['Age'].mean())
```
